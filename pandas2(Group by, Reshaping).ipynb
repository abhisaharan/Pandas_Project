# groupby

- Split a data frame into groups based on some criteria
- apply a function to each group indpendently
- combine the results into a data frame

```python
import pandas as pd
```

```python
olym_df = pd.read_csv('Data/SummerOlympics.csv', skiprows = 4)
olym_df.head()
```

# groupby object

```python
a = olym_df.groupby('Edition')
```

```python
list(a)
```

** Iterate through a group

```python
for key, value in olym_df.groupby('Edition'):
    print(key)
    print(value)
```

# group by computation

- agg(): multiple statistics in one calculation
- size(): print the total count of same row values in one column

```python
olym_df.groupby('Edition').size()
```

```python
olym_df.groupby(['Edition', 'NOC', 'Medal']).agg(['min', 'max', 'count'])
```

```python
olym_df.groupby(['Edition', 'NOC', 'Medal']).agg(['count'])
```

```python
olym_df.groupby(['Edition', 'NOC', 'Medal']).size()
```

** agg({:[ ...]}) : dict

```python
olym_df.groupby(['Edition', 'NOC', 'Medal']).agg({'Edition':['max', 'min','count']})
```

```python
olym_df[olym_df.Athlete == 'LEWIS, Carl'].groupby('Athlete').agg({'Edition':['min', 'max', 'count']})
```

# Reshaping: Stacking & unstacking

- stack(): help reshape dataframe
- unstack(): 



```python
mw = olym_df[(olym_df.Edition == 2008) & ((olym_df.Event == "100m") | (olym_df.Event == "200m"))]
mw
```

```python
g = mw.groupby(['NOC', 'Gender', 'Discipline', 'Event']).size()
g
```

```python
df = g.unstack(['Discipline', 'Event'])
df
```

** stack(): returns a df or series
- by default dropna = True i.e you will never see null values
- gives taller and longer df

```python
df.stack(['Event'])
```

** unstack(): 
- we can fill missing values ex: na to 0
- resulting df is wider

```python
df.unstack('Gender')
```

# challenge
- plot the number of gold medals won by the us male and female athletes throughout the history of the olympics?

- Plot the five athletes who have won the most gold medals over the history of the olympics. When there is a tie, consider the number of silver medals, and then bronze medals.

```python
usa = olym_df[(olym_df.NOC == 'USA') & (olym_df.Medal == 'Gold')]
usa.groupby(['Gender', 'Edition']).size().unstack('Gender', fill_value = 0).plot()
```

```python
most = olym_df.groupby(['Athlete', 'Medal']).size().unstack('Medal', fill_value = 0)
most.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold', 'Silver', 'Bronze']].head().plot(kind = 'bar')
```
