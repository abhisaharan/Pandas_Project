# Matplotlib

Bar graph: compare different groups. They are best when changes are large

Pie chart: they do not show changes over time. Used for comparison ofparts of whole

** colormap: Three classes of colormaps:
- sequential: used to represent info that has ordering
- Diverging: when info deviates around a middle value
- Qualitative: does not have any ordering or miscellaneous relationship

```python
fo.Sport.value_counts().plot(kind = 'pie', colormap = 'Pastel1' );
```

# Seaborn
** Seaborn: basic plotting lib based on matplotlib
- Its a compilment to matplotlib not a substitute.
- seaboard.countplot(data, hue, order, palette)
- data: Dataframe or source
hue: categorical values(one that can only take fixed no. of values)
order: sequence when using categorical levels
palette: colors to use for the different levels of the hue variable

Used for complex statistical plotting. Creates beautiful plots

```python
import seaborn as sns
```

```python
sns.countplot(x = 'Medal', data = olym_df, hue = 'Gender');
```
